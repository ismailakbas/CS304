```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

class_1_a = np.array([[4,2], [2,4], [2,3], [3,6], [4,4]])
class_2_a = np.array([[6,8], [9,5], [8,7], [10,8]])

class_1_b = np.array([[6,8], [2,4], [2,3], [3,6]])
class_2_b = np.array([[9,5], [8,7], [10,8], [4,2], [4,4]])

data_a = np.vstack((class_1_a, class_2_a))
labels_a = np.array([0]*5 + [1]*4)

data_b = np.vstack((class_1_b, class_2_b))
labels_b = np.array([0]*4 + [1]*5)

pca_a = PCA(n_components=2)
pca_b = PCA(n_components=2)

data_pca_a = pca_a.fit_transform(data_a)
data_pca_b = pca_b.fit_transform(data_b)

lda_a = LDA(n_components=1)
lda_b = LDA(n_components=1)

data_lda_a = lda_a.fit_transform(data_a, labels_a)
data_lda_b = lda_b.fit_transform(data_b, labels_b)

fig, ax = plt.subplots(2, 2, figsize=(12, 10))

ax[0, 0].scatter(class_1_a[:,0], class_1_a[:,1], color='purple', label='Class 1')
ax[0, 0].scatter(class_2_a[:,0], class_2_a[:,1], color='gold', label='Class 2')
ax[0, 0].set_title("PCA Vectors and Data Points for Dataset a ")
ax[0, 0].quiver(*np.mean(data_a, axis=0), *pca_a.components_[0], scale=3, color='black')

ax[1, 0].scatter(data_lda_a[labels_a==0], np.zeros_like(data_lda_a[labels_a==0]), color='purple', label='Class 1')
ax[1, 0].scatter(data_lda_a[labels_a==1], np.zeros_like(data_lda_a[labels_a==1]), color='gold', label='Class 2')
ax[1, 0].set_title("1D LDA Projection for Dataset a ")

ax[0, 1].scatter(class_1_b[:,0], class_1_b[:,1], color='purple', label='Class 1')
ax[0, 1].scatter(class_2_b[:,0], class_2_b[:,1], color='gold', label='Class 2')
ax[0, 1].set_title("PCA Vectors and Data Points for Dataset b")
ax[0, 1].quiver(*np.mean(data_b, axis=0), *pca_b.components_[0], scale=3, color='black')

ax[1, 1].scatter(data_lda_b[labels_b==0], np.zeros_like(data_lda_b[labels_b==0]), color='purple', label='Class 1')
ax[1, 1].scatter(data_lda_b[labels_b==1], np.zeros_like(data_lda_b[labels_b==1]), color='gold', label='Class 2')
ax[1, 1].set_title("1D LDA Projection for Dataset b")

for a in ax.flatten():
    a.legend()
    a.grid(True)

plt.tight_layout()
plt.show()
```
